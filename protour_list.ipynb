```python
import pandas as pd
import requests
import re
from bs4  import BeautifulSoup as BS
from time import sleep
from io   import StringIO
```

```python
from custom_status import print_refresh
```

```python
URL = 'https://en.wikipedia.org/wiki/List_of_Magic:_The_Gathering_Pro_Tour_events'

response = requests.get(URL)
print(response.status_code)
```

```python
soup = BS(response.text)
```

```python
tables = soup.findAll( 'table', attrs = {'class' : 'wikitable'} )
```

```python
protours_df = pd.read_html(
  StringIO( str( tables[1] ) )
)[0]
```

```python
inperson_df = protours_df[ 'Online' != protours_df['Location'] ]
```

```python
i = 0
standard_df = 0
for index, row in inperson_df.iterrows():
  if re.search( r'.*[sS]tandard.*', row['Format']) and 0==i :
    standard_df = row.to_frame() # reset_index(drop=True)
    i=+1
  elif re.search( r'.*[sS]tandard.*', row['Format']):
    standard_df = pd.concat([standard_df, row], axis=1)

standard_df = standard_df.transpose().reset_index(names='Pro Tour Number?')
```
